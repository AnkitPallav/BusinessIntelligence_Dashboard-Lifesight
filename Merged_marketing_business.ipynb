```python
import pandas as pd

# Load CSV files
business = pd.read_csv("business.csv")
tiktok = pd.read_csv("TikTok.csv")
google = pd.read_csv("Google.csv")
facebook = pd.read_csv("Facebook.csv")

# Preview data
business.head()
```

```python
# Aggregate TikTok by date
tiktok_daily = tiktok.groupby("date").agg({
    "impression": "sum",
    "clicks": "sum",
    "spend": "sum",
    "attributed revenue": "sum"
}).reset_index()
tiktok_daily.rename(columns={
    "impression": "tiktok_impression",
    "clicks": "tiktok_clicks",
    "spend": "tiktok_spend",
    "attributed revenue": "tiktok_revenue"
}, inplace=True)

# Repeat for Google & Facebook
google_daily = google.groupby("date").agg({
    "impression": "sum",
    "clicks": "sum",
    "spend": "sum",
    "attributed revenue": "sum"
}).reset_index()
google_daily.rename(columns={
    "impression": "google_impression",
    "clicks": "google_clicks",
    "spend": "google_spend",
    "attributed revenue": "google_revenue"
}, inplace=True)

facebook_daily = facebook.groupby("date").agg({
    "impression": "sum",
    "clicks": "sum",
    "spend": "sum",
    "attributed revenue": "sum"
}).reset_index()
facebook_daily.rename(columns={
    "impression": "fb_impression",
    "clicks": "fb_clicks",
    "spend": "fb_spend",
    "attributed revenue": "fb_revenue"
}, inplace=True)
```

```python
# Step 3: Merge all daily summaries with business data
merged_daily = business.merge(tiktok_daily, on="date", how="left") \
                       .merge(google_daily, on="date", how="left") \
                       .merge(facebook_daily, on="date", how="left")
```

```python
# Step 4: Save merged daily summary
merged_daily.to_csv("mergedMarketingData.csv", index=False)

# Optional: view top rows
print(merged_daily.head())
```
